```python
tts_model_path = '/content/glowtts.pth'
tts_conifg_path = '/content/glowtts_config.json'
vocoder_model_path = '/content/hifigan.pth.tar'
vocoder_conifg_path = '/content/hifigan_config.json'
```

## Speech Synthesis

```python
examples = ['গতকাল সোমবার পুরান ঢাকার চকবাজারসহ অন্যান্য এলাকা সরেজমিনে ঘুরে এবং স্থানীয় লোকজনের সঙ্গে কথা বলে চিত্র পাওয়া গেছে।',
            'ডলারের সংকট নিয়ে কোনো মন্তব্য করতে চাইছেন না ব্যাংকের ব্যবস্থাপনা পরিচালকেরা।',
            'বিদ্যুৎ সাশ্রয়ের জন্য অফিসের সময় এক ঘণ্টা কমাতে যাচ্ছে ঢাকা বিশ্ববিদ্যালয় কর্তৃপক্ষ৷',
            'রাশিয়ার ইউক্রেন আক্রমণের ছয় মাস হতে চলল, এখনো যুদ্ধ বন্ধের স্পষ্ট কোনো আভাস নেই।',
            'বাংলাদেশ সরকারের স্বাস্থ্য ও পরিবার কল্যাণ মন্ত্রণালয় দেশব্যাপী কিশোর ও প্রাপ্তবয়স্ক নাগরিকদের টিকাদান কার্যক্রমের সম্প্রসারণ অব্যাহত রাখতে পারে।']
```

```python
for i in range(len(examples)):
    output_path = 'example_output_'+str(i+1)+'.wav'
    subprocess.run(['tts', 
                    '--text', examples[i], 
                    '--model_path', tts_model_path, 
                    '--config_path', tts_conifg_path, 
                    '--out_path', output_path, 
                    '--vocoder_path', vocoder_model_path, 
                    '--vocoder_config_path', vocoder_conifg_path])
```
